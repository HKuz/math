# Trigonometry

**Trigonometry** is the branch in mathematics that studies the relationship between side lengths and angles in triangles. Trigonometric functions were originally only defined for acute right-angle triangles, but can be extended to a domain of all real numbers by using the unit circle (radius = 1).

Since trigonometry applies to triangles and circles, a quick refresher on the measure of angles in degrees and radians is below. Plots in the notebook will be in either Cartesian coordinates $(x, y)$ or Polar coordinates $(r, \theta)$.

**Degrees vs Radians**

Thousands of years ago, the Babylonians divided the path of the sun relative to the earth into 360ยบ. While this number is somewhat arbitrary, it has stood the test of time and is still used today.

Radians are more closely related to circles - one radian is defined as the angle corresponding to a portion of a circle with the same arc-length as the radius $r$. The unit circle demonstrates why $\pi$ shows up in radians - a unit circle has radius 1 and diameter 2. Since $\pi$ is defined as the ratio of a circle's circumference divided by its diameter $\pi = \frac{C}{D}$, an arc-length around the entire unit circle (which equals the circumference) is $\pi = \frac{C}{2}$, or $C = 2\pi$.

To convert between the two, use $\frac{\pi}{180^{\circ}} = \frac{\text{rad}}{\deg}$:

- Convert to degrees: $\deg = \frac{180 \cdot \text{rad}}{\pi}$
- Convert to radians: $\text{rad} = \frac{\deg \cdot \pi}{180}$


**Polar Coordinates**

Polar coordinates map a point $(r, \theta)$ and represent a system where each point on a plane is determined by a distance from a reference point and an angle from a reference direction ([Wikipedia](https://en.wikipedia.org/wiki/Polar_coordinate_system)).

You can convert polar coordinates to cartesian coordinates with the $\sin$ and $\cos$ functions:

- $x = r \cdot \cos(\theta)$
- $y = r \cdot \sin(\theta)$

## General Trigonometric Functions

Given a right triangle with sides $a$, $b$, and hypotenuse $h$, the basic trigonometric functions for $\sin(\theta)$, $\cos(\theta)$, and $\tan(\theta)$, plus their respective reciprocals $\csc(\theta)$ (cosecant), $\sec(\theta)$ (secant), and $\cot(\theta)$ (cotangent) are below. These functions are rooted in the Pythagorean theorem $a^2 + b^2 = h^2$. The functions are extended to a domain of all real numbers via the unit circle ($r = 1$), which is drawn in cartesian coordinates with the equation $x^2 + y^2 = r^2$, using $a$, $b$, and $h$ for $x$, $y$, and $r$, respectively.

The mnemonic SOH-CAH-TOA helps to remember the relationship of the given angle $\theta$ to the side-lengths of the right triangle, where O is the side opposite the angle, A is the side adjacent to the angle, and H is the hypotenuse.

- **Sine:** $\sin(\theta) = \frac{O}{H}$, and represents how the $y$ coordinate changes as $\theta$ moves a point around the unit circle
- **Cosine:** $\cos(\theta) = \frac{A}{H}$, and represents how the $x$ coordinate changes as $\theta$ moves a point around the unit circle
- **Tangent:** $\tan(\theta) = \frac{O}{A}$, and represents the ratio $\frac{y}{x}$ of the coordinates of a point for a given $\theta$ on the unit circle. Geometrically, the tangent is the length of a line segment tangent to that point $(x, y)$ on the unit circle at the given angle $\theta$, measured from the point $(x, y)$ to where it intersects the $x$-axis
- **Cosecant:** $\csc(\theta) = \frac{1}{\sin(\theta)} = \frac{H}{O}$, the reciprocal of the sine function
- **Secant:** $\sec(\theta) = \frac{1}{\cos(\theta)} = \frac{H}{A}$, the reciprocal of the cosine function
- **Cotangent:** $\cot(\theta) = \frac{1}{\tan(\theta)} = \frac{A}{O}$, the reciprocal of the tangent function. Geometrically, the cotangent is the length of the line segement tangent to the point $(x, y)$ on the unit circle at the given angle $\theta$, measured from the point $(x, y)$ to where it intersects the $y$-axis

```python
import math
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate data for sine and cosine plots
theta = np.linspace(-2 * math.pi, 2 * math.pi, 100)
s = np.sin(theta)
c = np.cos(theta)

# Plot the functions
plt.figure(figsize=(12, 6));
plt.plot(theta, s, 'r', label='$\sin$(rad)');
plt.plot(theta, c, 'b', label='$\cos$(rad)');
plt.title('Sine and Cosine Functions');
plt.xlabel('Radians');
plt.xticks([-2*math.pi, -3*math.pi/2, -math.pi, -math.pi/2, 0, math.pi/2, math.pi, 3*math.pi/2, 2*math.pi],
           ['$-2\pi$', '$-3\pi/2$', '$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']);
plt.axvline(x=0, c='lightgray');
plt.axhline(y=0, c='lightgray');
plt.legend();
```

```python
# Model for sine and cosine functions

def sin_pwr_series(theta, iters=12):
    """
    Returns the approximate value for sin(theta) for given angle theta
        using the power series definition
    :param theta: angle in radians
    :return: float
    
    >>>sin_pwr_series(math.pi / 2)
    1.0
    """
    y = 0
    for n in range(iters):
        y += ((-1)**n * theta**(2*n + 1)) / math.factorial(2*n + 1)
    return y

def cos_pwr_series(theta, iters=12):
    """
    Returns the approximate value for cos(theta) for given angle theta
        using the power series definition
    :param theta: angle in radians
    :return: float
    
    >>>cos_pwr_series(0)
    1.0
    """
    x = 0
    for n in range(iters):
        x += ((-1)**n * theta**(2*n)) / math.factorial(2*n)
    return x
```

```python
# Compare numpy implementation and power series for sin and cos
theta = math.pi / 4
test_sin = sin_pwr_series(theta)
test_cos = cos_pwr_series(theta)

print('Sin(x) comparison:')
print(f'Power series: {test_sin}')
print(f'Numpy sin(x): {np.sin(theta)}')

print()
print('Cos(x) comparison:')
print(f'Power series: {test_cos}')
print(f'Numpy sin(x): {np.cos(theta)}')
```

## Law of Sines and Law of Cosines

The utility of the sine and cosine functions extend beyond just right triangles - you can apply their definitions for right triangles in ways to also determine the angle and side-length relationships of any triangle. The below image shows a non-right triangle with angles $A$, $B$, and $C$ and corresponding side-lengths (opposite each angle) of $a$, $b$, and $c$, respectively. You can derive the **Law of Sines** and the **Law of Cosines** with this image in mind. Also, for the purposes of the calculations below, where $h$ intersects $b$, the left part of the split line segment will be called $x$ and the right part will be called $b - x$.



The **Law of Sines** states that $\frac{a}{\sin(A)} = \frac{b}{\sin(B)} = \frac{c}{\sin(C)}$. Taking the two right triangles created with $h$, the $\sin(A) = \frac{h}{c}$ and $\sin(C) = \frac{h}{a}$. Solving both for $h$ and setting them equal, you get $c \sin(A) = h = a \sin(C)$, and rearranging results in $\frac{a}{\sin(A)} = \frac{c}{\sin(C)}$. Rotating the triangle and dropping a new $h$ from the vertex of either $A$ or $C$ and following the same logic results in the final equality including $\frac{b}{\sin(B)}$.


The **Law of Cosines** states that $a^2 = b^2 + c^2 - 2bc \cos(A)$, which relates the lengths of each side to the cosine of one angle. To get here, first note that in the diagram above, $c^2 = h^2 + x^2$ (where $x$ is the left part of the split segment $b$) and $a^2 = h^2 + (b-x)^2$ (where $b-x$ is the right part of the split line segment $b$). Solving the first equation for $x^2$, so $x^2 = c^2 - h^2$ and substituting that into the second equation, you get $a^2 = b^2 -2bx + c^2$ (note that the $h^2$ terms cancel). To eliminate the final $x$ in the equation, note that $\cos(A) = \frac{x}{c}$, or $x = c\cos(A)$. You can substitute this for $x$, to get the final equation of $a^2 = b^2 + c^2 - 2bc\cos(A)$.

## Trigonometric Identities

Beyond the reciprocal functions noted above, there are several other useful identities.

### Even-Odd Identities

- $\sin(\theta)$ is an **odd function**, meaning that $\sin(\theta) = -\sin(-\theta)$, or more commonly: $\sin(-\theta) = -\sin(\theta)$
- $\cos(\theta)$ is an **even function**, meaning that $\cos(\theta) = \cos(-\theta)$
- $\tan(-\theta) = -\tan(\theta)$

### Cofunction Identities

- $\sin(\theta) = \cos(\frac{\pi}{2} - \theta)$
- $\cos(\theta) = \sin(\frac{\pi}{2} - \theta)$
- $\sec(\theta) = \csc(\frac{\pi}{2} - \theta)$
- $\csc(\theta) = \sec(\frac{\pi}{2} - \theta)$
- $\tan(\theta) = \cot(\frac{\pi}{2} - \theta)$
- $\cot(\theta) = \tan(\frac{\pi}{2} - \theta)$

### Pythagorean Identities

Given the unit circle, the hypotenuse of any right triangle created within it will be 1:

- $\sin^2(\theta) + \cos^2(\theta) = 1$
- Divide both sides by $\sin^2(\theta)$ to get $1 + \cot^2(\theta) = \csc^2(\theta)$
- Divide both sides by $\cos^2(\theta)$ to get $1 + \tan^2(\theta) = \sec^2(\theta)$

### Sum and Difference Identities

- $\sin(\alpha + \beta) = \sin(\alpha)\cos(\beta) + \cos(\alpha)\sin(\beta)$
- $\cos(\alpha + \beta) = \cos(\alpha)\cos(\beta) - \sin(\alpha)\sin(\beta)$
- $\sin(\alpha - \beta) = \sin(\alpha)\cos(\beta) - \cos(\alpha)\sin(\beta)$
- $\cos(\alpha - \beta) = \cos(\alpha)\cos(\beta) + \sin(\alpha)\sin(\beta)$

### Double Angle Identities

The sum identities can be used to derive the double angle identities, by using $(\theta + \theta)$.

- $\sin(2\theta) = 2\sin(\theta)\cos(\theta)$
- $\cos(2\theta) = \cos^2(\theta) - \sin^2(\theta)$

Substituting with the Pythagorean identity of $\sin^2(\theta) + \cos^2(\theta) = 1$, also get:

- $\cos(2\theta) = 1 - 2\sin^2(\theta)$
- $\cos(2\theta) = 2\cos^2(\theta) - 1$
